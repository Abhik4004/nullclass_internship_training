```python
img1 = cv2.imread('images/pngtree-warm-color-ocean-cartoon-crab-undersea-creature-png-image_328241.jpg')
plt.imshow(img1)
```

```python
img2 = cv2.imread('images/45984d90a07874bb10b7770215dfe212.jpg')
plt.imshow(img2)
```

```python
img = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
figure_size = 7
result = cv2.blur(img, (figure_size, figure_size))
plt.figure(figsize=(11, 6))
plt.subplot(121),plt.imshow(cv2.cvtColor(img, cv2.COLOR_HSV2RGB)),plt.title("original")
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(cv2.cvtColor(result, cv2.COLOR_HSV2BGR)), plt.title("Blurred")
plt.xticks([]),plt.yticks([])
plt.show()
```
